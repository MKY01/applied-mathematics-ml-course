## Exercise 1 Solution — Linear Regression

We fit a linear regression model and check the coefficients.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Simple dataset
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

model = LinearRegression().fit(X, y)

coef, intercept = model.coef_[0], model.intercept_
coef, intercept
```

**Explanation:**
- The fitted model is `y = 2x + 0`.
- Coefficient ≈ 2, Intercept ≈ 0, which matches the data perfectly.

## Exercise 2 Solution — Logistic Regression

We fit a logistic regression classifier on a small binary dataset.

```python
from sklearn.linear_model import LogisticRegression

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

clf = LogisticRegression().fit(X, y)

pred_probs = clf.predict_proba([[2.5], [4.5]])
pred_probs
```

**Explanation:**
- Logistic regression predicts probabilities.
- For input `2.5`, the model assigns higher probability to class 0.
- For input `4.5`, the model assigns higher probability to class 1.

## Exercise 3 Solution — Decision Tree Classifier

We fit a decision tree and visualize its predictions.

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

dt = DecisionTreeClassifier(max_depth=2).fit(X, y)

plt.figure(figsize=(6,4))
plot_tree(dt, filled=True, feature_names=["X"], class_names=["0","1"])
plt.show()
```

**Explanation:**
- Decision trees split data into regions based on feature thresholds.
- With `max_depth=2`, the tree learns a simple rule separating values ≤3 from values ≥4.

## Exercise 4 Solution — k-Nearest Neighbors

We apply kNN classification with `k=3`.

```python
from sklearn.neighbors import KNeighborsClassifier

X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

knn = KNeighborsClassifier(n_neighbors=3).fit(X, y)

knn.predict([[2.5]]), knn.predict([[5.5]])
```

**Explanation:**
- kNN classifies based on majority vote of nearest neighbors.
- At `2.5`, neighbors are mostly class 0 → prediction = 0.
- At `5.5`, neighbors are mostly class 1 → prediction = 1.
